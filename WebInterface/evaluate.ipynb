```python
queries = [
    {
    "query":"Military welfare department allowed for appointment to the existing post of clerk",
    "fromOrderID":"9/2022/MWDN"
    },
    {
    "query":"Sports equipment for financial year 2022",
    "fromOrderID":"253/2022/WaCDDN"},
    {
    "query":"additional post in Department of Energy",
    "fromOrderID":"8/2022/DoEN"
    },
    {
    "query":"Monsoon disaster in malappuram order released on 25/5/2022",
    "fromOrderID":"469/2022/DMDN"
    },
    {
    "query":"1339/2022/LSDN",
    "fromOrderID":None
    },
    {
    "query":"27/05/2022",
    "fromOrderID":None
    },
    {
    "query":"public work department",
    "fromOrderID":None
    },
    {
    "query":"department of public work",
    "fromOrderID":None
    }
    ,
    {
    "query":"order released on 27/05/2022",
    "fromOrderID":None
    }
    ]
```

```python
data = []
for element in queries:
    queryRespone = {}
    queryRespone["query"] = element["query"]
    queryRespone["fromOrderID"] = element["fromOrderID"]
    
    query = element["query"]
    oldMethodResults = SearchInterface(query,1,5,'BodyAndAbstract') #First five results of old method
    newMethodResults = SearchInterface(query,1,5,'AllAttributes') #First five results of new method
    resultsElement = []
    for result in oldMethodResults:
        resultData = {}
        resultData["ID"] = goid_formatter(result["GOID"])
        resultData["Date"] = str(result["Date"])
        resultData["Score"] = result["SCORE"]
        resultData["Department"] = result["Department"]
        resultsElement.append(resultData)
    queryRespone["oldMethodResults"] = resultsElement
    resultsElement = []
    for result in newMethodResults:
        resultData = {}
        resultData["ID"] = goid_formatter(result["GOID"])
        resultData["Date"] = str(result["Date"])
        resultData["Score"] = result["SCORE"]
        resultData["Department"] = result["Department"]
        resultsElement.append(resultData)
    queryRespone["newMethodResults"] = resultsElement
    # f.write(json.dumps(results, indent=4))
    # print(queryRespone)
    data.append(queryRespone)
jsonData = {v: k for v, k in enumerate(data)}
data
```

```python
mdTable = ""
for queryElement in data:
    query = queryElement["query"]
    fromOrderID = queryElement["fromOrderID"]
    oldMethodResults = queryElement["oldMethodResults"]
    newMethodResults = queryElement["newMethodResults"]
    mdOut = f"""||{query}<br>{fromOrderID}||\n| :---: | :---: | :---: |"""
    mdOut+= f"""\n|**Old Method**||**New Method**|"""
    for i in range(5):
        oldResultID = oldMethodResults[i]["ID"]
        newResultID = newMethodResults[i]["ID"]
        oldMethodDate = oldMethodResults[i]["Date"]
        newMethodDate = newMethodResults[i]["Date"]
        oldMethodScore = oldMethodResults[i]["Score"]
        newMethodScore = newMethodResults[i]["Score"]
        oldMethodDepartment = oldMethodResults[i]["Department"]
        newMethodDepartment = newMethodResults[i]["Department"]
        mdOut += f"""\n|{oldResultID}<br>{oldMethodDate}<br>{oldMethodDepartment}<br>{oldMethodScore}||{newResultID}<br>{newMethodDate}<br>{newMethodDepartment}<br>{newMethodScore}|"""
    mdTable += mdOut + "\n\n\n"
```

```python
f = open("compareTable.md","w")
f.write(mdTable)
```
